## Loading X and Y

```python
X = post_covid_ml_df.iloc[:, :-1].values
y = post_covid_ml_df.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
```

```python
print(y_pred.shape)
```

## Accuracy

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

print("r2:\n", r2_score(y_test, y_pred))
print("MAE:\n", mean_absolute_error(y_test, y_pred))
print("RMSE:\n", np.sqrt(mean_squared_error(y_test, y_pred)))
```

## Observation

```python
import statsmodels.api as sm
from statsmodels.stats.anova import anova_lm
from scipy import stats
```

### Metrics table

```python
residuals = y_test - y_pred

# number of observations
n = len(y_test)        

# number of predictors  
p = X_train.shape[1]     

# Sum of Squares
ss_total = np.sum((y_test - np.mean(y_test))**2)
ss_res = np.sum(residuals**2)
ss_model = ss_total - ss_res

# R-squared and adjusted R-squared
r2 = 1 - (ss_res / ss_total)
adj_r2 = 1 - (1 - r2) * ((n - 1) / (n - p - 1))

# Mean Square
ms_model = ss_model / p
ms_res = ss_res / (n - p - 1)

# F-statistic and p-value
F_stat = ms_model / ms_res
p_value = 1 - stats.f.cdf(F_stat, p, n - p - 1)

# Root MSE
root_mse = np.sqrt(ms_res)
```

#### Save metrics table

```python
metrics_table_df = pd.DataFrame({
    'Metric': [
        'Number of obs',
        f'F( {p}, {n - p - 1} )',
        'Prob > F',
        'R-squared',
        'Adj R-squared',
        'Root MSE'
    ],
    'Value': [
        n,
        round(F_stat, 2),
        f'{p_value:.4f}',
        round(r2, 4),
        round(adj_r2, 4),
        round(root_mse, 5)
    ]
})


os.makedirs("03-metrics-table", exist_ok=True)
metrics_table_df.to_csv("03-metrics-table/metrics_table.csv", index=False)
```

### ANOVA table

```python
n = len(y_test)
p = X_train.shape[1]

# Predicted and residuals
residuals = y_test - y_pred

# Sum of Squares
ss_total = np.sum((y_test - np.mean(y_test))**2)
ss_res = np.sum(residuals**2)
ss_model = ss_total - ss_res

# Degrees of freedom
df_model = p
df_resid = n - p - 1
df_total = n - 1

# Mean Squares
ms_model = ss_model / df_model
ms_resid = ss_res / df_resid
ms_total = ss_total / df_total
```

#### Save ANOVA table

```python
anova_table_df = pd.DataFrame({
    'Source': ['Model', 'Residual', 'Total'],
    'SS': [ss_model, ss_res, ss_total],
    'df': [df_model, df_resid, df_total],
    'MS': [ms_model, ms_resid, ms_total]
})

anova_table_df['SS'] = anova_table_df['SS'].map('{:.6f}'.format)
anova_table_df['MS'] = anova_table_df['MS'].map('{:.8f}'.format)

os.makedirs("04-anova-table", exist_ok=True)
anova_table_df.to_csv("04-anova-table/anova_table.csv", index=False)
```

### Coefficient Table

```python
# Convert to DataFrame with original column names
X_train_sm = sm.add_constant(pd.DataFrame(X_train, columns=post_covid_ml_df.columns[:-1]))

# Fit the model
model_sm = sm.OLS(y_train, X_train_sm).fit()

# Extract coefficients and metrics
params = model_sm.params
bse = model_sm.bse
tvalues = model_sm.tvalues
pvalues = model_sm.pvalues
conf = model_sm.conf_int()
```

#### Save coefficient table

```python
coeff_table_df = pd.DataFrame({
    'Variable': params.index,
    'Coef.': params.values,
    'Std. Err.': bse.values,
    't': tvalues.values,
    'P>|t|': pvalues.values,
    'Lower 95%': conf[0].values,
    'Upper 95%': conf[1].values
})

# Format numbers to match your screenshot
coeff_table_df['Coef.'] = coeff_table_df['Coef.'].map('{:.7f}'.format)
coeff_table_df['Std. Err.'] = coeff_table_df['Std. Err.'].map('{:.7f}'.format)
coeff_table_df['t'] = coeff_table_df['t'].map('{:.2f}'.format)
coeff_table_df['P>|t|'] = coeff_table_df['P>|t|'].map('{:.3f}'.format)
coeff_table_df['Lower 95%'] = coeff_table_df['Lower 95%'].map('{:.6f}'.format)
coeff_table_df['Upper 95%'] = coeff_table_df['Upper 95%'].map('{:.6f}'.format)

coeff_table_df.reset_index(drop=True, inplace=True)
print(coeff_table_df.head())


os.makedirs("05-coeff-table", exist_ok=True)
coeff_table_df.to_csv("05-coeff-table/coeff_table.csv", index=False)
```
