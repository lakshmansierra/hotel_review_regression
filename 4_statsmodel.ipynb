## filter post covid data with `dummies = 1`

```python
post_covid_df = selected_df[selected_df['dummies'] == 1].copy()
```

## drop dummies

```python
post_covid_ml_df = post_covid_df.copy()
```

```python
post_covid_ml_df.drop(columns=['dummies'], inplace=True)
```

## Loading X and Y

```python
X = post_covid_ml_df.iloc[:, :-1].values
y = post_covid_ml_df.iloc[:, -1].values
```

```python
print(X.shape)
print(y.shape)
```

## Observation

```python
# Add constant (intercept)
X_sm = sm.add_constant(X)

# Fit full regression model
model_full = sm.OLS(y, X_sm).fit()
```

### Metrics table

```python
n = int(model_full.nobs)
p = X.shape[1]
f_stat = model_full.fvalue
f_pvalue = model_full.f_pvalue
r2 = model_full.rsquared
adj_r2 = model_full.rsquared_adj
root_mse = np.sqrt(model_full.mse_resid)

metrics_table_df = pd.DataFrame({
    'Metric': [
        'Number of obs',
        f'F( {p}, {n - p - 1} )',
        'Prob > F',
        'R-squared',
        'Adj R-squared',
        'Root MSE'
    ],
    'Value': [
        n,
        round(f_stat, 2),
        f'{f_pvalue:.4f}',
        round(r2, 4),
        round(adj_r2, 4),
        round(root_mse, 5)
    ]
})
```

#### Save metrics table

```python
os.makedirs("06-metrics-table", exist_ok=True)
metrics_table_df.to_csv("06-metrics-table/06_metrics_table.csv", index=False)
```

### ANOVA table

```python
import statsmodels.formula.api as smf

dependent_var = post_covid_ml_df.columns[-1]
independent_vars = " + ".join(post_covid_ml_df.columns[:-1])
formula = f"{dependent_var} ~ {independent_vars}"

model_full = smf.ols(formula=formula, data=post_covid_ml_df).fit()
```

#### Model level ANOVA table

```python
anova_results = anova_lm(model_full, typ=1)

# Aggregate all predictors’ SS and df into a single “Model” entry
ss_model = anova_results.loc[anova_results.index != 'Residual', 'sum_sq'].sum()
df_model = anova_results.loc[anova_results.index != 'Residual', 'df'].sum()
ss_resid = anova_results.loc['Residual', 'sum_sq']
df_resid = anova_results.loc['Residual', 'df']
ss_total = ss_model + ss_resid
df_total = df_model + df_resid

# Compute Mean Squares (MS)
ms_model = ss_model / df_model
ms_resid = ss_resid / df_resid
ms_total = ss_total / df_total

# Build Stata-style table
anova_table = pd.DataFrame({
    'Source': ['Model', 'Residual', 'Total'],
    'SS': [f'{ss_model:.6f}', f'{ss_resid:.6f}', f'{ss_total:.6f}'],
    'df': [int(df_model), int(df_resid), int(df_total)],
    'MS': [f'{ms_model:.8f}', f'{ms_resid:.8f}', f'{ms_total:.8f}']
})
```

#### Save ANOVA table

```python
os.makedirs("07-anova-table", exist_ok=True)
anova_table.to_csv("07-anova-table/07_anova_table.csv", index=False)
```

### Coefficient Table

```python
params = model_full.params
bse = model_full.bse
tvalues = model_full.tvalues
pvalues = model_full.pvalues
conf = model_full.conf_int()

coeff_table_df = pd.DataFrame({
    'Variable': params.index,
    'Coef.': params.values,
    'Std. Err.': bse.values,
    't': tvalues.values,
    'P>|t|': pvalues.values,
    'Lower 95%': conf[0].values,
    'Upper 95%': conf[1].values
})

# Formatting like your image
coeff_table_df['Coef.'] = coeff_table_df['Coef.'].map('{:.7f}'.format)
coeff_table_df['Std. Err.'] = coeff_table_df['Std. Err.'].map('{:.7f}'.format)
coeff_table_df['t'] = coeff_table_df['t'].map('{:.2f}'.format)
coeff_table_df['P>|t|'] = coeff_table_df['P>|t|'].map('{:.3f}'.format)
coeff_table_df['Lower 95%'] = coeff_table_df['Lower 95%'].map('{:.6f}'.format)
coeff_table_df['Upper 95%'] = coeff_table_df['Upper 95%'].map('{:.6f}'.format)
coeff_table_df.loc[coeff_table_df['Variable'] == 'const', 'Variable'] = '_cons'
coeff_table_df.reset_index(drop=True, inplace=True)
```

#### Save coefficient table

```python
os.makedirs("08-coeff-table", exist_ok=True)
coeff_table_df.to_csv("08-coeff-table/08_coeff_table.csv", index=False)
```
